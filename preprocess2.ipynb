# *flatten the 2D array into 1D*

```python
import pandas as pd
import numpy as np


column_wise_array = train_crispr.T.values

# Reshape it into a 1D array
flattened_array = column_wise_array.reshape(-1)

# Create a new DataFrame with a single column
train_crispr_flatten = pd.DataFrame(flattened_array, columns=['Y'])

train_crispr_flatten.to_csv('/mnt/data/macaulay/datas/training_crispr.csv', index=False)
train_crispr_flatten
```
